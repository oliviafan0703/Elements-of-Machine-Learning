```python
import matplotlib.pyplot as plt
import numpy as np
import math 

def springs(k, a, phi, t):
    x_positions = []
    for time in t:
        x_positions.append(postion_at_instant(k,a,phi,time))
    return x_positions

def postion_at_instant(k, a, phi, t):
    position = []
    a0 = a[0]
    a1 = a[1]
    phi0 = phi[0]
    phi1 = phi[1]
    x0 = math.sqrt(2)/2 * a0 * math.cos(t + phi0) - math.sqrt(2)/2 * a1 * math.cos(math.sqrt(2*k+1) * t + phi1)
    x1 = math.sqrt(2)/2 * a0 * math.cos(t + phi0) + math.sqrt(2)/2 * a1 * math.cos(math.sqrt(2*k+1) * t + phi1)
    position =  [x0, x1]
    return position
```

```python
# 100 linearly spaced numbers
time = np.linspace(0,50,1000)

#pass the parameters
a = [1,3]
phi =[0, math.pi/3]
k = 5

# the function
x_positions = springs(k,a,phi,time)
t = []
x0_positions = []
x1_positions = []

n = 0
while n < len(time):
    t.append(time[n])
    x0_positions.append(x_positions[n][0])
    x1_positions.append(x_positions[n][1])
    n += 1

# plot the function
%matplotlib inline
plt.plot(t,x0_positions,label = "x0")
plt.plot(t,x1_positions,label = "x1")
plt.legend()


# show the plot
plt.xlabel('time')
plt.ylabel('deviation of position')
plt.title("deviation of position of each mass for k = 5")
plt.show()
```

### Problem 2.2

$$
p(0,n) = \int_{0}^{1} (1-h)^n \,dh \\
u=1-h,du=-dh\\
      = \int_{1}^{0} u^n \,-du \\
      = \int_{0}^{1} u^n \,du \\
      = \frac{u^{n+1}}{n+1} |_0^1\\
      = \frac{(1-h)^{n+1}}{n+1} |_1^0\\
      = \frac{1^{n+1}}{n+1} - \frac{0^{n+1}}{n+1}\\
      = \frac{1}{n+1}\\
p(n,n) = \int_{0}^{1} h^n \,dh \\
      = \frac{h^{n+1}}{n+1} |_0^1\\
      = \frac{1^{n+1}}{n+1} - \frac{0^{n+1}}{n+1}\\
      = \frac{1}{n+1}\\
$$

$p(0,n)= \int_0^1 p(0,n|h)\,\mathrm{d}h$
<br> $=  \int_0^1 (1-h)^n\,\mathrm{d}h$
<br> $= -\frac{1}{n+1}(1-h)^{n+1}|_0^1$
<br> $= \frac{1}{n+1}$

$p(n,n)= \int_0^1 p(n,n|h)\,\mathrm{d}h$
<br> $=  \int_0^1 h^n\,\mathrm{d}h$
<br> $= \frac{1}{n+1}h^{n+1}|_0^1$
<br> $= \frac{1}{n+1}$

### Problem 2.3

The cumulative distribution $ P_{h}(x|E_{n}) = P[h\leqslant
x|E_{n}] $
<br> $= \frac{P({h \leqslant x \cup E_{n}})}{P(E_{n})}$
<br> $= \frac {\int_0^x h^n\,\mathrm{d}h}
{P(E_{n})}$
<br> $= \frac{\frac{1}{n+1}h^{n+1}|_0^x}{\frac{1}{n+1}}$
<br> $=\frac{\frac{1}{n+1}x^{n+1}}{\frac{1}{n+1}}$
<br> $= x^{n+1}$

Taking derivative of the cumulative distribution,
<br> $ P_{h}(x|E_{n})= (n+1)x^n$

For $x=1,$ the probability density  $ P_{h}(x|E_{1})= 2x$
<br> For $x=2,$ the probability density  $ P_{h}(x|E_{2}) = 3x^2$
<br> For $x=3,$ the probability density  $ P_{h}(x|E_{3} = 4x^3$
<br> For $x=4,$ the probability density  $ P_{h}(x|E_{4})= 5x^4$
<br> For $x=5,$ the probability density  $ P_{h}(x|E_{5})= 6x^5$

```python
import matplotlib.pyplot as plt
import numpy as np

# 100 linearly spaced numbers
x = np.linspace(0,1,100)

# the function, which is y = 3x^2 here
y1 = 2*x
y2 = 3*x**2
y3 = 4*x**3
y4 = 5*x**4
y5 = 6*x**5

# plot the function
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.plot(x,y4)
plt.plot(x,y5)
plt.legend(["1","2","3","4","5"])


# show the plot
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title("Probability Density Distribution of $P_{h}(x|E_{n})$")
plt.show()
```

```python
def sample(n,m=100000):
    #random number generator
    rng = np.random.default_rng()
    #draw a large number distributed uniformly between 0 and 1.
    h = rng.uniform(0,1,m)
    #n trials, h probability
    bin = rng.binomial(n=n,p=h)
    #Retain and return values of h for which number drawn from binomial distribution is n
    #Rejection sampling as do not count the others
    out = h[bin == n]
    return out
    #rng.binomial
```

```python
n=1
while n < 6:
    my_bin = sample(n)
    plt.hist(x=my_bin, bins=100, range=(0,1),density=True)
    plt.title("n = "+str(n))
    plt.show()
    n+=2
```
